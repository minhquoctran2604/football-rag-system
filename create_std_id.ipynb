```python
# thêm tạo id chuẩn cho đội bóng dựa trên tên và năm thành lập
import json
import re
import unicodedata

output = './data/teams/team_complete_metadata_2023_2024.jsonl'

def slug(text: str) -> str:
    if not text:
        return ""
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    text = text.lower().strip()
    # thay mọi ký tự không phải a-z0-9 bằng '_'
    text = re.sub(r'[^a-z0-9]+', '_', text)
    return text.strip('_')


# with open('./data/teams/team_metadata_2023_2024.jsonl', 'r', encoding='utf-8') as teams:
#     count = 0
#     for line in teams:
#         team = json.loads(line)
#         team_name = slug(team['identity']['full_name'])
#         canonical_team_id = 'team_' + team_name + '_' + str(team['identity']['founded_year'])
#         team['canonical_team_id'] = canonical_team_id
#         with open(output, 'a', encoding='utf-8') as out:
#             out.write(json.dumps(team, ensure_ascii=False) + '\n')
```

```python
# link player -> team
team_index = dict()
with open('./data/teams/team_complete_metadata_2023_2024.jsonl','r',encoding='utf-8') as teams:
    for t in teams:
        t = json.loads(t)
        team_index[t.get("name",'UNKNOWN')] = t.get("canonical_team_id",'UNKNOWN')
        

with open('./data/players/players_complete_metadata_with_std_id.jsonl','r',encoding='utf-8') as players:
    for p in players:
        p = json.loads(p)
        current_club = slug(p.get("current_club","UNKNOWN"))
        if current_club in team_index:
            p['current_club_id'] = team_index[current_club]
            with open('./data/players/players_finals.jsonl','a',encoding='utf-8') as out:
                out.write(json.dumps(p, ensure_ascii=False) + '\n')
```
